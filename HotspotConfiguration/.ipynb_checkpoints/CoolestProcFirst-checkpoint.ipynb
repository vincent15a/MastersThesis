```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import subprocess

# Processor Specs
# Speed in GHz
proc_speed = 1.25

# Power Consumption in W
proc_power_low = 60
proc_power_high = 80
proc_power_idle = 5

# Processor Core Count is the number of processors per layer.
proc_layers = 2
proc_cores = 4

# Path to chip configuration
proc_config_path = "2-layer-chip"

# Location of the hotspot executable
hotspot = "/home/vincent/Desktop/HotSpot/hotspot"

nr_tasks = None
```

```python
data_path = "testdata.txt"
```

```python
task_graph = nx.DiGraph()
```

```python
f = open(data_path, 'r')
```

```python
task_power_consumptions = []
for line in f.readlines():
    line = line.strip().split()
    if line[0] == "v":
        task_graph.add_node(int(line[1]), weight=int(line[2]))
        task_power_consumptions.append(int(line[-1]))
    if line[0] == "e":
        task_graph.add_edge(int(line[2]), int(line[1]))
```

```python
def power_consumption(v):
    return task_power_consumptions[v]
```

```python
nx.draw(task_graph, with_labels=True)
plt.show()
```

```python
t = 0
```

```python
finished = set()
can_be_planned = set()
to_be_planned = set(task_graph.nodes)
running = list() # (task, end_time, PE)
nr_tasks = len(task_graph.nodes)
```

```python
def run_time(v):
    if not(0 <= v < nr_tasks):
        raise Exception("out of bounds")
    
    return task_graph.nodes[v]['weight']

def update_can_be_planned():
    for t in to_be_planned:
        plan = True
        for pred in task_graph.predecessors(t):
            if pred not in finished:
                plan = False
                break
        if plan:
            can_be_planned.add(t)

def get_temperatures_from_steadyfile(steady_file):
    # Read processor temperature from steady file.
    temps = []
    f = open(f'{proc_config_path}/{steady_file}', 'r')
    for line in f.readlines():
        line = line.split()
        if "PE" in line[0]:
            temps.append(float(line[-1]))
    f.close()
    return temps

def get_temperatures_from_simulation(transient, gridtransient):
    # Read processor temperature from transient and grid transient file.
    f_t = open(f"{proc_config_path}/{transient}", 'r')
    f_grid = open(f"{proc_config_path}/{gridtransient}", 'r')
    temps = list(map(float, f_t.readlines()[-1].split()))
    
    grid_file = f_grid.readlines()
    discard_before = 0
    for i in range(len(grid_file) - 1, -1, -1):
        if "t = " in grid_file[i]:
            discard_before = i
            break
    grid_file = grid_file[i:]
    
    TIM_temps = []
    i = 0
    while i < len(grid_file):
        if not "Layer " in grid_file[i]:
            i += 1
        else:
            line = grid_file[i].split()
            layer_nr = int(line[1][:-1])
            if layer_nr % 2 == 0 and layer_nr // 2 < proc_layers:
                i += 1
            else:
                lines = list(map(lambda x: float(x.split()[-1]), grid_file[i+1 : i+5]))
                TIM_temps.append(sum(lines)/len(lines))
                i += 1
    f_t.close()
    f_grid.close()
    return (temps, TIM_temps)

def write_temperatures_to_init(temps, TIM_temps, output_file_name):
    f = open(f'{proc_config_path}/{output_file_name}', 'w')
    for layer in range(proc_layers * 2):
        if layer % 2 == 0:
            for i in range(proc_cores):
                f.write(f"layer_{layer}_PE_L{layer//2 + 1}_{i + 1} \t {temps[layer//2 * proc_cores + i]}\n")
        else:
            f.write(f"layer_{layer}_TIM_1 \t {TIM_temps[layer // 2]}\n")
    f.write(f"hsp_TIM_1 \t {TIM_temps[-2]}\n")
    f.write(f"hsink_TIM_1 \t {TIM_temps[-1]}\n")
    f.close()
        

def construct_hotspot_command(power_trace, config=f'example.config', grid_layer_file=f"layers.lcf", steady_file=None, grid_steady_file=None, output=None, grid_output=None):
    command = [hotspot, "-c", config, "-p", power_trace, "-grid_layer_file", grid_layer_file, "-model_type", "grid", "-detailed_3D", "on"]
    if grid_steady_file != None:
        command = command + ["-grid_steady_file", grid_steady_file]
    if steady_file != None:
        command = command + ["-steady_file", steady_file]
    if output != None:
        command = command + ["-o", output]
    if grid_output != None:
        command = command + ["-grid_transient_file", grid_output]
    return command

def write_power_trace(power_trace_lines, output):
    f = open(f'{proc_config_path}/{output}', 'w')
    for i in range(1, proc_layers + 1):
        for j in range(1, proc_cores + 1):
            f.write(f"PE_L{i}_{j} ")
    f.write("\n")
    
    for line in power_trace_lines:
        if len(line) != proc_layers * proc_cores:
            raise Exception("Wrong power trace file!")
        for nr in line:
            f.write(f"{nr} ")
        f.write("\n")
```

```python
update_can_be_planned()
```

```python
# Set initial steady temperatures
def set_initial_temps(output, grid_output):
    f = open(f'{proc_config_path}/initial.ptrace', 'w')
    for i in range(1, proc_layers + 1):
        for j in range(1, proc_cores + 1):
            f.write(f"PE_L{i}_{j} ")
    f.write("\n")
    
    for _ in range(100):
        for i in range(proc_layers * proc_cores):
            f.write(f"{proc_power_idle} ")
        f.write("\n")
    f.close()
    
    subprocess.run(construct_hotspot_command(f'initial.ptrace', output=output, grid_output=grid_output), cwd=proc_config_path)
```

```python
# Here we perform the actual simulation, with as strategy planning the next task on the coldest core.

# Set intial steady file.
set_initial_temps("prev_temps.transient", "prev_grid.ttrace")
temps, TIMtemps = get_temperatures_from_simulation("prev_temps.transient", "prev_grid.ttrace")
write_temperatures_to_init(temps, TIMtemps, "prev_temps.init")

# Current time
time = 0

while len(finished) != nr_tasks:
    t = can_be_planned.pop()
    to_be_planned.remove(t)
    
    # If there are tasks with weight 0, we do not actually simulate them
    # These are added as the first / last task from datasets.
    if run_time(t) == 0:
        finished.add(t)
        update_can_be_planned()
        continue
    
    
    # Check current time, see if there are tasks in running that should go
    for item in running:
        task, end_time, PE = item
        if end_time <= time:
            running.remove(item)
    
    occupied_cores = [p for (_, _, p) in running]
    free_cores = [p for p in range(proc_layers * proc_cores) if p not in occupied_cores]
    
    temps, _ = get_temperatures_from_simulation("prev_temps.transient", "prev_grid.ttrace")
    
    min_free_core = min(free_cores, key=lambda x: temps[x])
    
    running.append((t, time + run_time(t), min_free_core))
    
    running.sort(key=lambda x: x[-1]) # Sort by PE
    update_can_be_planned()
    run_time_steps = []
    # Check if the simulation has to be ran, or that we can plan more
    if len(to_be_planned) == 0:
        max_running_time = max(map(lambda x: x[1], running)) # Find out how many time steps to simulate
        while t != max_running_time:
            run_time_step = []
            for (task, end_time, _) in running:
                if time < end_time:
                    run_time_step.append(power_consumption(task))
                else:
                    run_time_step.append(proc_power_idle)
            t += 1
            run_time_steps.append(run_time_step)
        write_power_trace(run_time_steps, 'simulate.ptrace')
    
    elif len(running) == proc_cores * proc_layers:
        min_running_time = min(map(lambda x: x[1], running)) # Find out when a task finishes first
        nr_of_timesteps = min_running_time - time
        
        run_time_step = []
        for (task, _, _) in running:
            run_time_step.append(power_consumption(task))
        
        run_time_steps = [run_time_step] * nr_of_timesteps
        write_power_trace(run_time_steps, "simulate.ptrace")
        
        t += nr_of_timesteps
    
    # Run simulation if needed
    if len(run_time_steps) != 0:
        subprocess.run(construct_hotspot_command('simulate.ptrace', output='prev_temps.transient', grid_output='prev_grid.ttrace'), cwd=proc_config_path)
```
